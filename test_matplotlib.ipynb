```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-3, 3, 50)

y = 2*x + 1
y2 = x**2
plt.figure()
l1, = plt.plot(x,y,label = 'linear line')
l2,= plt.plot(x,y2,color='red',linewidth=2.0,linestyle='--',label = 'square line')
plt.xlim((-1,2))
plt.ylim((-2,5))
new_ticks = np.linspace(-1,2,10)
print(new_ticks)
plt.xticks(new_ticks)
plt.yticks([-1,0,1],['china','orange','banana'])
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.xaxis.set_ticks_position('bottom') #top bottom default
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('default')  #left right default
ax.spines['left'].set_position(('data',0))
plt.legend(loc = 'best') #plt.legend(loc = 'upper left')
plt.legend(handles = [l1,l2],labels =['z','m'], loc = 'best')

#annotation
x0 = 1
y0 = 2*x0 +1
plt.plot([x0,x0],[0,y0],'k--', linewidth = 2.5)
plt.scatter([x0,],[y0,], color = 'b',s = 50)  # s is the size of dot

plt.annotate('2x+1=%s' %y0, xy=(x0,y0), xycoords='data',textcoords = 'offset points',
             xytext = (-30,30),fontsize=16, arrowprops=dict(arrowstyle='->',connectionstyle='arc,rad=.2'))

plt.show()
```
